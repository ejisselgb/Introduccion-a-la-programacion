### Caso 1

#### Secuenciación del ADN y ARN

El ácido desoxirribonucleico y ácido ribonucleico corresponden a los ácidos nucleicos, los cuales son un tipo importante de macromoléculas, las cuales están presentes en todas las células y virus.  
El ácido desoxirribonucleico o ADN es el portador de información genética y a su vez se encarga de replicar dicha información, mientras que el ácido ribonucleico o ARN se produce a partir del ADN.  

La estructura primaria de los ácidos nucleicos está dada por una secuencia de nucleótidos, conformada por:  

- Puricas: Adenina, Guanina
- Pirimidinas: Citosina, Timina, Uracilo

El ADN es posible encontrarlo en el núcleo y mitocrondia de la célula. El ARN se encuentra en el citoplasma.

Para formar cadenas de ADN o ARN, se requiere la unión de varios nucleotidos, a lo que se le conoce como secuenciación de ADN.  

La secuenciación de ADN permite determinar el orden de los 4 componentes básicos llamados "bases", para así establecer la clase de información genética que se transporta en un segmento específico. Los datos de las secuencias resaltan los cambios de un gen que pueden por ejemplo causar enfermedades.



Las bases para cada tipo de ácido son:

- **ADN**:  

(A) adenina  
(G) guanina  
(C) citosina  
(T) timina


- **ARN**:  

(A) adenina  
(G) guanina  
(C) citosina  
(U) uracilo


Un ejemplo de una secuenciación de una cadena de ADN es:

- CTTACGAACGTA (ADN)  
- GAAUGCUUGCAU (ARN)

Con el fin de disminuir tiempo en el proceso de secuenciación de diversas cadenas de ADN tomadas por algunos estudiantes de Biología para estudiar la información genética de múltiples organismos vivos y detectar posibles cambios y problemas en los genes, se deberá crear un programa que lleve a cabo la secuenciación de una cadena de ADN.
