### ¿Qué es un lenguaje de programación?

Conjunto de instrucciones que permite crear programas para controlar el comportamiento físico y lógico de una máquina. El programa puede dar órdenes, comparar valores, procesar y almacenar datos.  

Un lenguaje de programación posee un conjunto de reglas semánticas que le dan orden y estructura, además permite que una persona pueda comunicarse con una máquina por medio de símbolos y palabras claves.

### Tipos de lenguajes de programación

En la programación existen varios tipos de lenguajes, los cuales han sido desarrollados para cumplir un objetivo o tarea específica. Pueden estar destinados a aplicaciones para análisis de datos, inteligencia artificial, programación de controladores y hardware, aplicaciones web o móviles.  
Los tipos de lenguajes se presentan a continuación:

### Bajo nivel

El código de bajo nivel trabaja a gran velocidad, es de fácil conversión para la máquina, pero puede ser complejo para el programador escribirlo, dado que su sintaxis está muy ligada al lenguaje del hardware.

#### Código Binario  

Es un lenguaje básico conformado por los números 0 y 1. Permite procesar las instrucciones de una computadora, dadas desde código binario o por medio de un lenguaje de programación (interpretado o compilado), el cual su traducción se reduce a bits para que sea posible la lectura desde un computador. (más bajo nivel)


#### Lenguaje máquina

Lenguaje de programación diseñado para cada máquina, permitía la comunicación directa, pero era complejo construir código a través de este, pues era necesario aprender uno diferente para cada dispositivo. El código binario es lenguaje de máquina.

#### Lenguaje emsamblador

Lenguaje que utiliza códigos o abreviaturas que podían ser memorizadas, sin embargo requiere un traductor que convierta los programas construidos en este lenguaje a lenguaje de máquina. Fue el primer intento de creación de un lenguaje que fuera comprendido por los programadores pero que a su vez pudiera ser comprendido por la máquina a través de un intermediario.


### Alto nivel

Permiten la construcción de código a través de instrucciones dadas por medio del lenguaje que comúnmente utilizamos, estas pueden ser en inglés, español, etc. Los lenguajes de alto nivel no son comprendidos por la máquina por lo que deben pasar por un traductor o un compilador.

#### Traductor o intérprete:

Convierte el código de alto nivel y lo traduce a lenguaje de máquina para que este pueda ser comprendido. Mientras este se traduce, se va ejecutando y la máquina va realizando las instrucciones.

#### Compilador

Permite traducir todo un programa diseñado en un lenguaje de alto nivel a un lenguaje de máquina. Esta traducción es almacenada, por lo que no requiere una traducción por cada ejecución a no ser de que presente cambios.




### ¿Qué es python?


Es un lenguaje de programación imperativo, de sintaxis limpia y de código libre. Es un lenguaje de alto nivel, el cual puede ser utilizado en diversas aplicaciones web, inteligencia artificial, analítica de datos y videojuegos.  

### Instalando python

1. Ir al siguiente enlace para comenzar la descarga: https://www.python.org/downloads/
2. Seleccionar la opción download en el menú superior
3. Presionar botón de descargar python 3.9.6 para iniciar la descarga
4. Iniciar el ejecutable y selecionar las opciones ''Install launcher for all users'' y ''add python 3.7 to PATH''

5. Presionar el botón install now
6. Abrir consola del computador y ejecutar el comando `python`
