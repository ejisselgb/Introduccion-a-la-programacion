## Patrones repetitivos

Al momento de escribir código es muy común encontrarnos con elementos que se repiten una y otra vez, por ejemplo una operación matemática o la manera en que se captura un dato y se muestra un mensaje. Es importante identificar dentro de nuestro código los bloques que se repiten, para así a partir de muchos bloques crear uno solo, que ejecute la misma funcionalidad muchas veces. 

Cuando se optimiza el código y se eliminan elementos repetidos, pasando de tener muchas lineas de código a unas pocas, también se le llama refactorizar, haciendolo más fácil de entender, recuerda que siempre se debe programar pensando en el otro, ya que nuestro código puede ser manipulado por personas ajenas a nosotros.

#### Principios del reúso de código:

- Reusabilidad: El reúso de código hace referencia a volver en un único bloque un fragmento que se repite muchas veces dentro de nuestro código. Esto evitará tener código repetido y que la modificación sea más fácil.  

- Modularidad:  Hace referencia a crear bloques, módulos o funciones que agrupen ciertos fragmentos en funciones específicas.

Un programa bien escrito debe cumplir con ciertas características, estas beben ser:

- **Compacto** no necesariamente largo
- **Legible** fácil de entender y utilizar por otros
- **Robusto** No fallar con entradas inesperadas
- **Reutilizable** Se pueden utilizar bloques del programa para nuevos módulos o componentes o otros programas

En los lenguajes de programación existen las **funciones** las cuales permiten lograr el reúso de código. Las funciones encapsulan bloques de código que luego pueden ser llamados desde cualquier lugar del código y ser nuevamente utilizados sin escribir todas las lineas de código que hay en su interior.   

Recuerda el uso de `len()` o `input()`son funciones propias del lenguaje de programación que nos permiten conocer el tamaño de un texto o capturar datos desde la consola por parte del usuario.

## Funciones

Como bien ya vimos, las funciones permiten hacer reúso de código, las partes que la componen son:

- Nombre
- Argumentos de entrada
- Bloque de código
- Parámetros de salida

```python
# Escribiendo funciones en Python

def suma_funcion(num1, num2): #se debe poner def al inicio para saber que es una función. luego se le asigna un nombre
    respuesta = num1 + num2 #Se realiza una operación con los datos de entrada
    print("El resultado de la suma es: " + str(respuesta))

suma_funcion(2,5)
```

### Tipos de funciones

Una función puede ser declarada con:

- **Parámetros**, recibe valores para ser operados al interior de la función.
- **Sin parámetros**, No recibe valores y puede operar con valores internos o globales del programa.

```python
## Ejemplo de función sin parametros

print("Saludar usuario")
nombre= input("Ingrese el nombre del usuario: ")

def saludo():
    print("Bienvenido: " + nombre)

saludo()
```

### Sentencia Return

El uso de `return`permitira salir de una función y transferir la ejecución de vuelta a donde se realizó la llamada y a su vez devolver parámetros, fruto de la ejecución de la función.

**Importante: El código escrito después de un return en una función no es ejecutado**

```python
## Función con return

def multiplicacion(a=3,b=3):
    respuesta= a*b
    return respuesta
    
funcion_multi = multiplicacion()
print("El resultado de la multiplicacion es ", funcion_multi)
```
