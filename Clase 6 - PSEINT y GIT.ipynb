## Uso de GIT

**Para la ejecución de esta práctica es necesario tener instalado GIT https://git-scm.com/downloads**

Para el almacenamiento de nuestros trabajos de código, utilizaremos la herramienta GIT, esta nos permite manipular nuestro código de manera que puedo devolverme en el tiempo por si hay un problema con el código actual, ver quién aplicó cambios, cuales fueron las últimas modificaciones y estado de cada modificación. Para iniciar vamos a ir al enlace:  

https://github.com/ejisselgb/Introduccion_Programacion_Practica.git  

Y seguido de esto vamos a realizar lo siguiente:

- **git clone** https://github.com/ejisselgb/Introduccion_Programacion_Practica.git, esta url se obtiene del recuadro tal como se muestra en la imagen



Ya teniendo nuestro proyecto de manera local (el proyecto en gitgub se le conoce como proyecto remoto o en la nube), vamos ingresar a la ubicación de la carpeta desde la consola y vamos a revisar las ramas del proyecto con el siguiente comando:

- **git branch -a**

Una rama en git es una sección que se crea de una rama principal (main o master), donde es posible realizar cambios de un código sin afectar el de los demás, todo lo que ocurre en la rama creada desde la principal, no tiene relevancia en las demás hasta que el desarrollador o programador se lo indique.

Para crear una nueva rama ejecutaremos el siguiente comando:

- **git branch new_branch** o **git checkout -b new_branch**
new_branch será el nombre personalizado por cada uno, en donde la estructura será nombre_apellido, por lo que quedaría
- **git branch erika_gutierrez** o **git checkout -b erika_gutierrez**

Para subir los cambios realizados a la rama ejecutamos los siguientes comandos:

- **git add .** para agregar todas las modificaciones sobre la rama o **git add path_file** para agregar solo un archivo en particular al proyecto de manera remota
- **git commit -m "message"** agregar un mensaje a los cambios, funciona como un break point o un identificador en la línea de tiempo
- **git push origin name_branch** o **git push** para subir los cambios a la rama creada, estos ya podrán visualizarse en el repositorio de Github

Para descargar cambios de la rama principal o para descargar cambios que se han aplicado sobre nuestra rama pero no la tenemos de manera local ejecutar

- **git pull origin master** para descargar cambios de la rama principal, estos ya por defecto quedarían en mi rama o **git pull** para descargar solo los cambios de mi rama
