### Pseudocódigo

Un pseudocódigo es la manera de representar un algoritmo ya sea en lenguaje natural o en representaciones gráficas a través
de diagramas de flujo. Un pseudocódigo también puede escribirse bajo la sintaxis propia de un lenguaje de programación.
La construcción de un algoritmo por medio de pseudocódigo facilita el paso a un lenguaje de programación, para la creación de un programa.

#### Ejemplo de algoritmo en pseudocódigo



#### Representación en lenguaje de programación Python

```python
print("Ingresa el primer número de la suma")
numero1 = int(input("numero1: "))
print("Ingresa el segundo número de la suma")
numero2 = int(input("numero2: "))

sumar = numero1 + numero2
print("el resultado de la suma es: ")
print(sumar)
```

**La dieferencia entre un lenguaje de programación y un pseudocódigo, es que el pseudocódigo es un lenguaje "falso" que utiliza lenguaje natural para la descripción de sus pasos, mientras que un lenguaje de programación utiliza su propia sintaxis y realiza tareas de mayor complejidad.** 

En pseudocódigo también podemos utilizar los operadores matemáticos que se utilizan en un lenguaje de programación convencional como:

- Suma (+)
- Resta (-)
- Multiplicación (*)
- División (/)
- Potencia (^)
- Modulo (%)

También es posible utilizar los comparadores matemáticos

- Mayor que (>)
- Menor que (<)
- Igual (=)
- Menor o igual (<=)
- Mayor o igual (>=)

Generalmente, los diagramas de flujo, permiten hacer una migración a un lenguaje de cpnstrucción de algoritmos mediante pseudocódigo, hay elementos equivalente que facilitan sus transcripción y entendimiento.




La estructura del pseudocódigo es:

1. Nombre del algoritmo a construir
2. Bloque de declaración de identificadores, que se utilizan en la ejecución de un programa
3. Resultado del procedimiento

Para la construción de pseudocódigo y la validación del mismo, existen herramientas como **PSEINT** la cual ofrece un entorno de programación y ejecución del código escrito, simulando un ambiente real de programación, permitiendo un mejor entendimiento de la lógica de programación.

Los comandos vigentes en el entorno de programación en pseudocódigo se presentan a continuación:

 

<center>Este comando permite mostrar un mensaje o un resultado final en la pantalla al usuario</center>


<center>Este comando permite capturar un valor por consola, insertado por un usuario</center>


<center>Este comando permite almacenar un valor en una variable o constante</center>


<center>Este comando permite realizar validaciones, similares a las condiciones de los diagramas de flujo</center>


<center>Es similar a las validaciones de los diagramas, sin embargo lo que hace es comprobar que una variable contiene un valor determinado y según su valor realizar la opción asociada</center>


<center>Este comando ejecuta un conjunto de instrucciones mientras una sentencia sea correcta o se cumpla</center>


<center>Este comando repite una instrucción hasta se le indique en la setencia el valor de parada de la ejecución</center>


<center>Encapsula instrucciones repetidas que pueden ser reutilizadas en otros algoritmos</center>
