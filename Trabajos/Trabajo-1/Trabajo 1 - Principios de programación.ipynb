Al tiempo 0 y=x  
Después   
1 generación y = x.2  
2 generaciones y = (x.2)2=(2^2)x  
n generaciones y = (2^n)x  

n lo que corresponde al número de generaciones se calcula usando la siguiente ecuación  


log y = log x + n log 2
n = logy - logx / log2

log2 tiene un valor de 0.3010 y 1/0.3010 = 3.3  

por lo que n = 3.3 log y/x


La ecuación anterior permite encontrar el número de generaciones que han tenido lugar. Se deben conocer los valores para X y Y.

### Caso 2: Ajuste de código en python

Hoy en día, es muy común que los dispositivos electrónicos cuenten con sistema de geolocalización. Los dispositivos tienen dos formas de obtener la información de una ubicación ya sea directamente por latitud y longitud o por grados, minutos y segundos, a esto último se le conoce como coordenadas decimales.

En el caso de las coordenadas cartesianas corresponden al punto X y Y de un mapa.






**Ejemplo de conversión de coordenadas cartesianas a coordenadas decimales **


Según la información de la latitud y longitud entregada:

Latitud = 6.252574
Longitud = -75.565096


Se obtiene:

Latitud 

Grados 6.252574
Minutos 0.252574 * 60 = 15.15444
Segundos 0.15444 * 60 = 9.2664


Coordenadas decimales para la latitud 6°15'09.3''N


Longitud  

Grados -75.565096
Minutos 0.565096 * 60 = 33.90576
Segundos 0.90576 * 60 = 54.3456

Coordenadas decimales para la longitud 75°33'54.4''W

El programa entregado captura la información del GPS en forma de latitud y longitud, y debe realizar la conversión a grados, minutos y segundos, pero sus validaciones está incompletas y adicionalmente hay errores en el código entregado. Solucione los errores del código y valide la respuesta entregada en google maps. Permita que el valor de la latitud y longitud sea pedida a través de la consola. Adjunte capturas a su respuesa

```python
import math

latitud = 6.252574

grados_dd, grados = math.modf(latitud)
grados = str(float(grados))
minutos = grados_dd * 40
minutos_dd, minuto = math.modf(minutos)
segundos = minutos_dd * 60
segundos = round(float(segundos),2)


if(segundos > 10):
    segundos = "0"+str(segundos)


print("La latitud en coordenadas decimales es: " + grados + "°" + str(int(minuto)) + "'" + str(segundos) + "''" + "N")


```
