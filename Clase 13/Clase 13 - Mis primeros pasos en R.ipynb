## ¿Qué es R?

R es un lenguaje y un entorno para gráficos y computación estadística. R proporciona una amplia variedad de técnica estadísticas (modelo lineal y no lineal, pruebas estadísticas clásicas, análisis de series de tiempo, clasificación y agrupamiento), es altamente extensible. 

R está disponible bajo la licencia de software libre, y puede ser instalado en múltiples plataformas MacOS, Windows y Lunix.

#### El entorno de R

R a nivel de software cuenta con un conjunto integrado de paquetes y módulos para la manipulación de datos, cáculos y visualización de gráficas permitiendo:

1. Instalación eficaz de manejo y almacenamiento de datos
2. Conjunto de operadores para cálculos en matrices
3. Amplia colección coherente e integrada de herramientas intermedias para el análisis de datos
4. Facilidades gráficas para el análisis de datos y visualización en pantalla o en papel

**R se puede vincular con lenguajes de programación C o C++ para realizar tareas de computación intesiva**

## Características de R

R cuenta con una colección de herramientas para el manejo, almacenamiento y operaciones con datos. R es un lenguaje de programación simple pero potente.

R debe ser utilizado cuando se procesarán grandes cantidades de datos cuando están desordenados, diferentes tipos de datos en un mismo archivo, datos grandes y complejos, varias operaciones para procesar los datos que son geográficos.

## Conceptos básicos

- **Entorno:** Espacio de trabajo donde se edita y se crea software
- **Workspace:** Espacio aislado ¿donde el programador puede realizar sus actividades
- **Comando:** Directiva que permite realizar una acción u operación
- **Operador:** Ejecuta una acción asociada a un cálculo específico
- **Compilar:** Código interprado por una máquina y es ejecutado

## Introducción a R

Para comprender mejor R y algunos términos a utilizar, considerando que se hará uso de una interfaz gráfica, para llevar a cabo la programación en R, se definen las siguientes palabras, utilizadas propiamente en el área de la programación.

- **¿Qué es un GUI? (Graphical user interface)**: Programa que permite interactuar con un sistema de manera gráfica, evitando así, trabajar desde consola.
- **¿Qué es un IDE? (Integrated development enviroment)**: Programa que facilita el desarrollo de código, constando de editor, compilación y depuración

**R Studio** es un IDE con GUI diseñado para facilitar la escritura y visualización de software diseñado en R

Las ventajas de utilizar herramientas de apoyo y gráficas para programar son:

- Facilidad en la escritura de código
- Facilidad de visualización del código
- Facilidad de acceso de donde se almacenan los datos y librerías
- Facilidad en la publicación y visulización de los resultados

#### Archivos compatibles

Algunos de los archivos compatibles en R para la importación y tratamiento de datos son: `xls, csv, sql, html, txt`(aquí se nombran los más conocidos)

**Importante**: Un archivo compaatible con R, no necesariamente podrá ser manipulado en este lenguaje

#### Formatos de archivos creados en R

Los formatos de los archivos que se crean por medio del IDE de R son de tipo `.R`. Otros archivos serán de tipo `.Rdata`. Estos tienen respaldo de información, acelera el proceso en R guarda datos y objetos temporales que son de utilidad la momento de desarrollar.


## Configuración básica de R studio

Teniendo en cuenta los conceptos previos de entorno y workspace vistos previamente, se definen estos nuevamente, ya que al inicio en que construye el proyecto, estos elementos son los que primero son creados. 

- **Workspace:** Permite trabajar sobre un espacio temporal, guardando elementos, características y comportamientos (definiciones, variables). Todo esto se almacena sobre el archivo de respaldo .Rdata. Sin embargo el workspace puede ser almacenado cuando se guarda el proyecto de manera global, lo que evita que los datos temporales se pierdan.

- **Directorio:** Carpetas donde se almacenan los archivos ya sean multimedia o de código.

Cuando se crea un proyecto en R, se recomienda almacenar los archivos en carpetas, teniendo en cuenta la siguiente estructura

`Carpeta principal -> datos -> codigo -> resultados`

##### Para tener en cuenta

- Para crear archivos en R, ir a la opción new file en el panel derecho, luego new file. Es importante escoger la ubicación del archivo en un directorio conocido.
- La interfaz puede personalizarse utilizando la opción del menu superior `Tools -> Global tools -> appearence`
- Definir siempre las ubicaciones, para que las modificaciones del códigio queden siempre en el mismo lugar
- El botón source ejecuta el código del script de R en su totalidad
- El botón Run, ejecuta únicamente la linea sobre la cual se encuentra el cursor
