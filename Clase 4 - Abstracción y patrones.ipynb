#### Implicancia o condicional (⇒ / →)

*p* entonces *q*, es falsa solo si **q** es falsa, en las demás situaciones es verdadera, esto permite realizar demostraciones por contradicción

| p | q | → | r |
|----| --- | --- | ---|
| True | True | --> | True |  
| True | False | --> | False |
| False | True | --> |True |
| False | False | --> | True |


#### Equivalencia o bicondicional (⇔ / ↔)

*p* sy y sólo si *q*, es verdadera solo si **p** y **q** tienen el mismo valor (verdadero - verdadero / falso - falso), en las demás situaciones es verdadera, esto permite realizar demostraciones por contradicción

| p | q | ↔ | r |
|----| --- | --- | ---|
| True | True | --> | True |  
| True | False | --> | False |
| False | True | --> |False |
| False | False | --> | True |

## Reconocimiento de patrones

### Patrones simples:

Los patrones simples están asociadas a las características. Consisten en una secuencia de elementos los cuales se repiten en el mismo orden. Para identificar patrones simples el procedimiento a realizar es:

- Buscar sustantivos que aparecen repetidamente. Estos pueden ser objetos/propiedades (variables) usados en el código .
- Buscar verbos que aparecen repetidamente. Estos pueden ser operaciones o funciones
- Buscar descripciones específicas. Estas podrían ser abstraidas por terminos (variables) más generales. Por ejemplo, (rojo, grande, suave) --> (color, tamaño, textura)


### Patrones complejos:

Los patrones complejos corresponden a instrucciones que pueden ser generalizadas en:

- Ciclos
- Subrutinas, funciones
- Reglas, condicionales, ecuaciones

## Abstracción

En programación la abstración consiste en centrarse en el funcionamiento de una aplicación, se enfoca en el *qué* hace más no en *cómo* lo hace. 

Abstracción está muy relacionado al concepto de **caja negra**, donde se desconocen implementaciones complejas las cuales proporcionan funcionalidades faciles y sencillas de realzar para una persona. 

Piensa en un elemento de la vida cotidiana que ofrezca un conjunto de herramientas que te hagan más fácil de realizar algunas actividades, pero de las cuales desconoces sus funcionamiento interno. 

Cada funcionamiento puede ser componente abstracto que se representa en la vida real, por ejemplo un Televisor, tiene funcionalidades sencillas como apagar y encender, desconoces el proceso interno para llevar a cabo dicho proceso, pero accionar el evento es muy sencillo desde un control remoto.
